```python
import folium
import json
```

```python
with open('ncr_geojson_4326.geojson', 'r') as geojson_file:
    ncr_geojson = json.load(geojson_file)

with open('ncr_typography.json', 'r') as json_file:
    ncr_typography = json.load(json_file)

for city in ncr_geojson["features"]:
    city_name = city["properties"]["adm3_en"]
    city_properties = city["properties"]
    city_properties["trivia"] = ncr_typography[city_name]
```

```python
m = folium.Map(location=[14.594180245262951, 121.02610069274301],
               zoom_start=11,
               min_zoom=11,
               tiles="CartoDB.PositronNoLabels"
    )

tooltip = folium.GeoJsonTooltip(
    fields=['adm3_en', "trivia"],
    aliases=["Name", ""],
    localize=True,
    sticky=False,
    labels=True,
    style=  """
            background-color: #F0EFEF;
            border: 2px solid black;
            border-radius: 3px;
            box-shadow: 3px;
            min-width: 200px;
            max-width: 500px;
            white-space: normal; 
            word-wrap: break-word;  
            """,
    )

 
folium.GeoJson(ncr_geojson,
               style_function=lambda x: {'color':'black','fillColor':'transparent','weight':1.5},
               tooltip=tooltip,
               highlight_function=lambda x: {'weight':3,'fillColor':'grey'},
              ).add_to(m)

m
```

```python
m.save("index.html")
```

1. Import NCR files.

4. Style as desired
